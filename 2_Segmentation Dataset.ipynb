```python
# Check what was downloaded
from pathlib import Path

cholecseg_path = Path("cholecseg8k_raw")  # Convert string to Path object

if cholecseg_path.exists():
    items = list(cholecseg_path.rglob('*'))
    files = [x for x in items if x.is_file()]
    folders = [x for x in items if x.is_dir()]
    
    print(f"\n📊 Downloaded content:")
    print(f"  Total files: {len(files)}")
    print(f"  Total folders: {len(folders)}")
    
    # Show structure
    print(f"\n📂 Directory structure (first 20 folders):")
    for folder in sorted(folders)[:20]:
        print(f"  - {folder.relative_to(cholecseg_path)}")
    
    print(f"\n📄 Sample files (first 20):")
    for file in sorted(files)[:20]:
        size_mb = file.stat().st_size / (1024**2)
        print(f"  - {file.relative_to(cholecseg_path)} ({size_mb:.2f} MB)")
    
    # Analyze file types
    from collections import defaultdict
    extensions = defaultdict(int)
    for file in files:
        ext = file.suffix.lower()
        extensions[ext] += 1
    
    print(f"\n📊 File types:")
    for ext, count in sorted(extensions.items(), key=lambda x: x[1], reverse=True):
        print(f"  {ext if ext else 'no extension'}: {count} files")
        
else:
    print("❌ Download path not found!")
```

```python
# Analyze CholecSeg8k structure
from pathlib import Path
import re

cholecseg_path = Path("cholecseg8k_raw")

# Get all video folders
video_folders = sorted([f for f in cholecseg_path.iterdir() if f.is_dir() and f.name.startswith('video')])

print(f"📹 CholecSeg8k Dataset Structure")
print(f"="*60)
print(f"Total videos: {len(video_folders)}")

# Analyze each video
video_stats = []
for video_folder in video_folders[:5]:  # First 5 videos
    # Get sequence folders (like video01_00080)
    sequences = sorted([f for f in video_folder.iterdir() if f.is_dir()])
    
    total_frames = 0
    for seq in sequences:
        frames = len(list(seq.glob('*_endo.png')))  # Count original frames
        total_frames += frames
    
    video_stats.append({
        'name': video_folder.name,
        'sequences': len(sequences),
        'frames': total_frames
    })
    
    print(f"\n📁 {video_folder.name}")
    print(f"   Sequences: {len(sequences)}")
    print(f"   Annotated frames: {total_frames}")
    if sequences:
        print(f"   Example: {sequences[0].name}")

print(f"\n{'='*60}")
print(f"✅ CholecSeg8k downloaded successfully!")
print(f"   - Segmentation masks for surgical instrument detection")
print(f"   - Multiple mask types (binary, color, watershed)")
print(f"   - Ready for blood loss estimation training")
```
